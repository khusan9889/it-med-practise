```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv('sub-division_population_of_pakistan.csv')
df.head(10)
```

```python
#Calculate the rural and urban men
total_male_rural = df['MALE (RURAL)'].sum()
total_male_urban = df['MALE (URBAN)'].sum()
print(total_male_rural)
print(total_male_urban)
```

```python
#Calculate the rural and urban women
total_female_rural = df['FEMALE (RURAL)'].sum()
total_female_urban = df['FEMALE (URBAN)'].sum()
print(total_female_rural)
print(total_female_urban)
```

```python
#Calculate total population in rural and urban
total_population_rural = df['POPULATION 1998 (RURAL)'].sum()
total_population_urban = df['POPULATION 1998 (URBAN)'].sum()
print(total_population_rural)
print(total_population_urban)
```

```python
#Calculate the population in rural areas and create new column
population_rural = df['Total_population_(RURAL)'] = total_female_rural + total_male_rural
population_urban = df['Total_population_(URBAN)'] = total_female_urban + total_male_urban
total_population = df['Total_population'] = population_rural + population_urban
df.head(5)
```

```python
percent_male_rural = (total_male_rural / population_rural) * 100
percent_female_rural = (total_female_rural / population_rural) * 100
print(f"Male percentage living in rural areas: {percent_male_rural}%")
print(f"Female percentage living in rural areas: {percent_female_rural}%")
```

```python
#Data of pie chart
labels = ['Male', 'Female']
sizes = [percent_male_rural, percent_female_rural]
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90, colors=['blue', 'pink'], wedgeprops=dict(width=0.4))
plt.title('Population Distribution in Rural Areas by Gender')
plt.show()
```

```python
percent_male_urban = (total_male_urban / population_urban) * 100
percent_female_urban = (total_female_urban / population_urban) * 100
print(f"Male percentage living in urban areas: {percent_male_urban}%")
print(f"Female percentage living in urban areas: {percent_female_urban}%")
```

```python
#Data of pie chart
labels = ['Male', 'Female']
sizes = [percent_male_urban, percent_female_urban]
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90, colors=['blue', 'pink'], wedgeprops=dict(width=0.4))
plt.title('Population Distribution in Urban Areas by Gender')
plt.show()
```
